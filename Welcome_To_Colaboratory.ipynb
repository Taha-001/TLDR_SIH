```
import pickle
```

```
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM
```

```
mt5_tokens=AutoTokenizer.from_pretrained("csebuetnlp/mT5_multilingual_XLSum")
mt5_model=AutoModelForSeq2SeqLM.from_pretrained("csebuetnlp/mT5_multilingual_XLSum")
```

```
pickle.dump(mt5_model,open("mt5_model.pkl","wb"))
pickle.dump(mt5_tokens,open("mt5_tokenizer.pkl","wb"))
```

```
tokens=mt5_tokens.encode(data,truncation=True,return_tensors="pt")
```

```
summary=mt5_model.generate(tokens)
```
